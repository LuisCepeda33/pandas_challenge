## Accessing rows and values

These are working examples. Use brackets when you only want columns. Use `loc` or `iloc` when you want rows and columns together.

### Brackets for simple column selection

Brackets are usually the clearest choice when you want all rows but only a few columns.

```python
orders[['order_id', 'customer_id', 'status']].head(4)
```

### Filtering with brackets

You can also filter rows with brackets by putting a condition inside the first set of brackets, then optionally selecting columns with a second set.

```python
orders[orders['total_amount'] >= 200][['order_id', 'customer_id', 'status', 'total_amount']].head()
```

### `loc` for label-based row and column selection

Use `loc` when you want to choose rows and columns in one step by their labels.

```python
orders.loc[0:3, ['order_id', 'customer_id', 'status']]
```

### `iloc` for position-based row and column selection

Use `iloc` when you want the first few rows or columns by numeric position instead of by label.

```python
orders.iloc[0:4, 0:4]
```

## Setting values with `loc` and `iloc`

Use a copy when you are experimenting so the original data stays unchanged.

```python
orders_demo = orders[['order_id', 'status', 'shipping_city']].head(5).copy()
orders_demo.loc[0, 'status'] = 'Reviewed'
orders_demo.iloc[1, 2] = 'Updated City'
orders_demo
```

## Sorting and selecting columns

```python
orders[['order_id', 'status', 'total_amount']].sort_values('total_amount', ascending=False).head()
```
